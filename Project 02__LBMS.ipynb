```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter as tk
import tkinter.messagebox as tkMassgeBox

```

```python
fullframe=Tk()
fullframe.title(' Project Library Book Managment System')

def database():
    
    global conn,cursorobj
    conn=sqlite3.connect('BOOK.db')
    cursorobj=conn.cursor()
    cursorobj.execute("CREATE TABLE IF NOT EXISTS Datatable(authorname TEXT,bookname TEXT,publishedyear INTEGER,isbn INTEGER)")
    

def display():
    
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM Datatable ORDER BY bookname ASC")
    fetch=cursorobj.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    cursorobj.close()
    conn.close()
    tk.messagebox.showinfo('Library','Data fetched click ok')
    
def search():
    title_author=((authorname.get()).title())
    title_bookname=((bookname.get()).title())
    title_isbn=((isbn.get()).title())
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM Datatable")
    fetch=cursorobj.fetchall()
    count=0
    for data in fetch:
        if(title_author==data[0]) and (title_bookname==data[1]) and (str(publishedyear.get())==str(data[2])) and (str(title_isbn)==str(data[3])):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(title_author==data[0]):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(title_bookname==data[1]):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(str(publishedyear.get())==str(data[2])):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
        elif(str(title_isbn)==str(data[3])):
            tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
            count+=1
    if(count==0):
        tk.messagebox.showinfo('Library','No book like this in library')
    cursorobj.close()
    conn.close()
    authorname.set("")
    bookname.set("")
    publishedyear.set("")
    isbn.set("")
    
    
    
    
    
def addbook():
    
    title_author=((authorname.get()).title())
    title_bookname=((bookname.get()).title())
    title_isbn=((isbn.get()).title())
    if ((authorname.get()=='') or (bookname.get()=='') or (publishedyear.get()=='') or (isbn.get()=='')):
        tk.messagebox.showinfo('Library','Please fill details')
    else:
        database()
        cursorobj.execute("INSERT INTO Datatable(authorname,bookname,publishedyear,isbn) VALUES(?,?,?,?)",(title_author,title_bookname,publishedyear.get(),title_isbn))
        conn.commit()
        authorname.set("")
        bookname.set("")
        publishedyear.set("")
        isbn.set("")
        cursorobj.close()
        conn.close()
        tk.messagebox.showinfo('Library','Values inserted')
        
def delete():
    
    title_author=((authorname.get()).title())
    title_bookname=((bookname.get()).title())
    title_isbn=((isbn.get()).title())
    title_publishedyear=(publishedyear.get().title())
    tree.delete(*tree.get_children())
    database()
    cursorobj.execute("SELECT * FROM Datatable")
    fetch=cursorobj.fetchall()
    for data in fetch:
        if (title_author==data[0]):
            temp=("'"+str(title_author)+"'")
            temp1="DELETE FROM Datatable WHERE authorname= "+temp
            cursorobj.execute(temp1)
            tk.messagebox.showinfo('Library','Row is deleted')
            conn.commit()
            break
        elif(title_bookname==data[1]):
            temp=("'"+str(title_bookname)+"'")
            temp1="DELETE FROM Datatable WHERE bookname = "+temp
            cursorobj.execute(temp1)
            tk.messagebox.showinfo('Library','Row is deleted')
            conn.commit()
            break
        elif(str(title_isbn)==str(data[3])):
            temp=("'"+str(title_isbn)+"'")
            temp1="DELETE FROM Datatable WHERE isbn = "+temp
            cursorobj.execute(temp1)
            tk.messagebox.showinfo('Library','Row is deleted')
            conn.commit()
            break
        elif(str(title_publishedyear)==str(data[2])):
            temp=("'"+str(title_publishedyear)+"'")
            temp1="DELETE FROM Datatable WHERE publishedyear = "+temp
            cursorobj.execute(temp1)
            tk.messagebox.showinfo('Library','Row is deleted')
            conn.commit()
            break
            
    else:
        tk.messagebox.showinfo('Library','No data with this name')
    authorname.set("")
    bookname.set("")
    publishedyear.set("")
    isbn.set("")
    conn.commit()
    cursorobj.close()
    conn.close()
    
    
def issue():
    
        title_author=((authorname.get()).title())
        title_bookname=((bookname.get()).title())
        title_isbn=((isbn.get()).title())
        title_publishedyear=(publishedyear.get().title())
        tree.delete(*tree.get_children())
        database()
        cursorobj.execute("SELECT * FROM Datatable")
        fetch=cursorobj.fetchall()
        cursorobj.execute("DELETE FROM Datatable")
        count=0
        for data in fetch:
            if(title_author==data[0]) and (title_bookname==data[1]) and (title_publishedyear==str(data[2])) and (title_isbn==str(data[3])):
                if(count==0):
                    count+=1
                    tk.messagebox.showinfo('Library','Book issued')
                else:
                    cursorobj.execute("INSERT INTO Datatable(authorname,bookname,publishedyear,isbn) VALUES(?,?,?,?)",(data[0],data[1],data[2],data[3]))
                    conn.commit()
            else:
                cursorobj.execute("INSERT INTO Datatable(authorname,bookname,publishedyear,isbn) VALUES(?,?,?,?)",(data[0],data[1],data[2],data[3]))
                conn.commit()
        if(count==0):
            tk.messagebox.showinfo('Library','No book with detail in library')
            authorname.set("")
            bookname.set("")
            publishedyear.set("")
            isbn.set("")
            conn.commit()
            cursorobj.close()
            conn.close()
    
    
    
    
    
def Exit():
    result=tkMassgeBox.askquestion("Do you want to exit? (y/n)",icon="warning")
    if result=='yes':
        fullframe.destroy()
        exit()

        
        
        
        
        
##Frames

TitleFrame=Frame(fullframe,width=1100,padx=18,bd=20,relief=RIDGE)
TitleFrame.pack(side=TOP)
lblTitle=Label(TitleFrame,width=45,font=('arial',40,'bold'),text='# Library Book Managment System #',padx=10)
lblTitle.grid()



ButtonFrame=Frame(fullframe,bd=20,padx=18,relief=RIDGE)
ButtonFrame.pack(side=BOTTOM)

LeftFrame=Frame(fullframe,bd=20,width=500,height=350,padx=18,relief=RIDGE)
LeftFrame.pack(side=LEFT)

Form=Frame(LeftFrame,bd=20,width=500,height=350,padx=18)
Form.pack()

RightFrame=Frame(fullframe,bd=20,width=600,height=350,padx=18,relief=RIDGE)
RightFrame.pack(side=RIGHT)

##Variables

authorname=StringVar()
bookname=StringVar()
publishedyear=StringVar()
isbn=StringVar()

##Label and Entry

lbl_authorname=Label(Form,text="Author Name:",font=('arial',15,'bold'))
lbl_authorname.grid(row=0,column=0,sticky='w')
entry_authorname=Entry(Form,textvariable=authorname,width=30)
entry_authorname.grid(row=0,column=1)

lbl_bookname=Label(Form,text="Book Name:",font=('arial',15,'bold'))
lbl_bookname.grid(row=1,column=0,sticky='w')
entry_bookname=Entry(Form,textvariable=bookname,width=30)
entry_bookname.grid(row=1,column=1)

lbl_publishyear=Label(Form,text="Published year:",font=('arial',15,'bold'))
lbl_publishyear.grid(row=2,column=0,sticky='w')
entry_publishyear=Entry(Form,textvariable=publishedyear,width=30)
entry_publishyear.grid(row=2,column=1)

lbl_ISBN=Label(Form,text="ISBN No:",font=('arial',15,'bold'))
lbl_ISBN.grid(row=3,column=0,sticky='w')
entry_ISBN=Entry(Form,textvariable=isbn,width=30)
entry_ISBN.grid(row=3,column=1)


##Buttons

but_view=Button(ButtonFrame,text="View Library",width=10,font=('arial',10,'bold'),command=display)
but_view.grid(row=0,column=0)

but_search=Button(ButtonFrame,text="Search",width=10,font=('arial',10,'bold'),command=search)
but_search.grid(row=0,column=1)

but_add=Button(ButtonFrame,text="Add Book",width=10,font=('arial',10,'bold'),command=addbook)
but_add.grid(row=0,column=2)

but_issue=Button(ButtonFrame,text="Issue Book",width=10,font=('arial',10,'bold'),command=issue)
but_issue.grid(row=0,column=3)

but_delete=Button(ButtonFrame,text="Delete Book",width=10,font=('arial',10,'bold'),command=delete)
but_delete.grid(row=0,column=4)

but_exit=Button(ButtonFrame,text="Exit",width=10,font=('arial',10,'bold'),command=Exit)
but_exit.grid(row=0,column=5)



##Scrollbar

scrollbary=Scrollbar(RightFrame)
scrollbarx=Scrollbar(RightFrame)

col=('AUTHOR_NAME','BOOK_NAME','PUBLISHED_YEAR','ISBN')
tree=ttk.Treeview(RightFrame,columns=col,show='headings')
scrollbary=Scrollbar(RightFrame)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx=Scrollbar(RightFrame)
scrollbarx.pack(side=BOTTOM,fill=X)


tree.heading("AUTHOR_NAME",text="AUTHOR NAME",anchor=W)
tree.heading("BOOK_NAME",text="BOOK NAME",anchor=W)
tree.heading("PUBLISHED_YEAR",text="PUBLISHED YEAR",anchor=W)
tree.heading("ISBN",text="ISBN",anchor=W)

tree.column("AUTHOR_NAME",minwidth=0,width=180)
tree.column("BOOK_NAME",minwidth=0,width=180)
tree.column("PUBLISHED_YEAR",minwidth=0,width=300)
tree.column("ISBN",minwidth=0,width=180)
tree.pack(side=TOP,fill=BOTH)


fullframe.mainloop()
```
